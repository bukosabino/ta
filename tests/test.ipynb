# Ploting different new features

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import sys
sys.path.append("..") # Adds higher directory to python modules path.
from ta import *
```

```python
# Load data
df = pd.read_csv('../data/datas.csv', sep=',')
df = utils.dropna(df)
```

```python
df.head()
```

```python
df = add_all_ta_features(df, "Open", "High", "Low", "Close", "Volume_BTC")
df = df.fillna(method='backfill')
```

# Ploting some volatility features

### Average True Range

```python
plt.plot(df[40500:41000].Close)
plt.plot(df[40500:41000].volatility1)
plt.title('Average True Range')
plt.show()
```

### Bollinger Bands

```python
plt.plot(df[40500:41000].Close)
plt.plot(df[40500:41000].volatility2)
plt.plot(df[40500:41000].volatility3)
plt.plot(df[40500:41000].volatility4)
plt.title('Bollinger Bands')
plt.show()
```

### Keltner Channel

```python
plt.plot(df[40500: 41000].Close)
plt.plot(df[40500: 41000].volatility7)
plt.title('Keltner Channel')
plt.show()
```

### Donchian Channel

```python
plt.plot(df[40500: 41000].Close)
plt.plot(df[40500: 41000].volatility8)
plt.plot(df[40500: 41000].volatility9)
plt.title('Donchian Channel')
plt.show()
```
