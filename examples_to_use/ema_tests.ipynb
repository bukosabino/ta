```python
import pandas as pd
import ta


df = pd.read_csv('../data/datas.csv', sep=',')

# Clean nan values
df = ta.utils.dropna(df)

```

```python
n_fast=12

df['sma'] = df.Close.rolling(n_fast).mean()
df['emafast1'] = df.Close.ewm(n_fast).mean()

sma = df.Close.rolling(window=n_fast, min_periods=n_fast).mean()[:n_fast]
rest = df.Close[n_fast:]
df['emafast2'] = pd.concat([sma, rest]).ewm(span=n_fast, adjust=False).mean()

```

```python
import matplotlib.pyplot as plt
df[['Close', 'sma', 'emafast1', 'emafast2']].iloc[-20:,:].plot(figsize=(12, 10));
plt.show()
```
